    الگوریتم <i>تحلیل مولفه‌های اصلی</i> با نام <code>PCA</code> در پکیج <code>decomposition</code> این کتابخانه در دسترس است. برخی از آرگومان‌های مهم آن در جدول زیر آمده است، اما جهت مطالعه‌ی کامل‌تر مستندات می‌توانید به <a href="https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html" target="_blank">این لینک</a> مراجعه فرمایید.
    <br>
    برای استفاده از این الگوریتم، می‌توانید با اجرای <code>from sklearn.decomposition import PCA</code> آن را وارد (<code>import</code>) کنید.

<center>
<div dir=rtl style="direction: rtl;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>

|آرگومان|جنس و تایپ|توضیحات|
|:------:|:--------:|:---:|
|n_components|<code>int</code> یا <code>float</code>|تعداد مولفه‌هایی است که قصد داریم پس از اجرای الگوریتم، آن‌ها را داشته باشیم و حذف نشوند. اگر به این آرگومان مقدار ندهیم، همه مولفه‌ها حفظ خواهند شد. اگر جنس این آرگومان <code>int</code> باشد، نشانگر تعداد مولفه‌هایی است که قصد داریم حفظ شود. اما اگر از جنس <code>float</code> و به شکل $0< n\_components< 1$ باشد، به این معنا خواهد بود که به اندازه‌ای از مولفه‌ها حفظ شوند که واریانس داده‌ها بزرگتر از `n_componets`٪ باشد. بنابراین می‌توان اینگونه برداشت کرد که هرچه این عدد به ۱ نزدیک‌تر باشد، چون قرار است داده واریانس زیادی داشته باشد، پس مولفه‌های بیشتری حفظ می‌شوند.|
    
</font>
</div>
</center>

<div dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">

|اسم متد|توضیحات|
|:------:|:---:|
|fit(X)|الگوریتم را با X که در ابعاد `(n_samples, n_features)` است|
|transform(X)|این تابع X که از ابعاد `(n_samples, n_features)` است را به یک ماتریس دیگری با ابعاد `(n_samples, n_components)` تبدیل می‌کند. پس خروجی این تابع یک ماتریس به ابعاد `(n_samples, n_components)` خواهد بود|

```python
# Preprocessing
```

```python
# Preprocessing - کاهش ابعاد با PCA
# ابتدا با حفظ 95% واریانس شروع می‌کنیم
pca = PCA(n_components=0.95, random_state=42)

# آموزش PCA روی داده‌های آموزشی
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print(f"تعداد مؤلفه‌های انتخاب شده: {pca.n_components_}")
print(f"شکل جدید داده‌های آموزشی: {X_train_pca.shape}")
```

```python
# تقسیم داده‌های آموزشی به train و validation
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_pca, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"تعداد نمونه‌های آموزشی: {len(X_train_split)}")
print(f"تعداد نمونه‌های اعتبارسنجی: {len(X_val_split)}")
```

```python
# آموزش مدل SVM با پارامترهای بهینه
model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)

# آموزش مدل
print("در حال آموزش مدل SVM...")
model.fit(X_train_split, y_train_split)
print("آموزش مدل تکمیل شد!")
```

```python
# Evaluation
# پیش‌بینی روی داده‌های اعتبارسنجی
y_val_pred = model.predict(X_val_split)

# محاسبه f1_score
f1_val = f1_score(y_val_split, y_val_pred, pos_label='Real', average='binary')
print(f"F1-Score روی داده‌های اعتبارسنجی: {f1_val:.4f}")

# نمایش گزارش کامل
print("\nگزارش دسته‌بندی:")
print(classification_report(y_val_split, y_val_pred))
```

```python
import numpy as np
import pandas as pd
import scipy.sparse as ss

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, f1_score
from sklearn.preprocessing import StandardScaler

# 1. خواندن داده‌ها
train_data = pd.read_csv('news_train.csv')
train_data_text_vectors = ss.load_npz('news_train_text_vectors.npz')

test_data = pd.read_csv('news_test.csv')
test_data_text_vectors = ss.load_npz('news_test_text_vectors.npz')

# تبدیل ماتریس sparse به dense
X_train_full = train_data_text_vectors.toarray()
X_test = test_data_text_vectors.toarray()

# استخراج برچسب‌ها
y_train_full = train_data['label'].values

print(f"شکل داده‌های آموزشی: {X_train_full.shape}")
print(f"شکل داده‌های تست: {X_test.shape}")
print(f"توزیع برچسب‌ها: {np.unique(y_train_full, return_counts=True)}")

# 2. پیش‌پردازش - استانداردسازی
scaler = StandardScaler(with_mean=False)
X_train_scaled = scaler.fit_transform(X_train_full)
X_test_scaled = scaler.transform(X_test)

# 3. کاهش ابعاد با PCA
# ابتدا با حفظ 99% واریانس شروع می‌کنیم
pca = PCA(n_components=0.99, random_state=42)
print("در حال اعمال PCA...")
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

print(f"تعداد مؤلفه‌های انتخاب شده: {pca.n_components_}")
print(f"شکل جدید داده‌های آموزشی: {X_train_pca.shape}")

# 4. تقسیم داده‌ها برای اعتبارسنجی
X_train, X_val, y_train, y_val = train_test_split(
    X_train_pca, y_train_full,
    test_size=0.2,
    random_state=42,
    stratify=y_train_full
)

# 5. آموزش مدل SVM با پارامترهای بهینه‌شده
# بر اساس نتایج ضعیف اولیه، از پارامترهای بهتری استفاده می‌کنیم
model = SVC(
    kernel='rbf',
    C=10.0,  # افزایش C برای مدل قوی‌تر
    gamma='auto',  # استفاده از gamma خودکار
    random_state=42,
    class_weight='balanced'  # مدیریت عدم تعادل کلاس‌ها
)

print("در حال آموزش مدل SVM...")
model.fit(X_train, y_train)
print("آموزش مدل تکمیل شد!")

# 6. ارزیابی مدل
y_val_pred = model.predict(X_val)

# محاسبه f1_score برای کلاس Real
f1_val = f1_score(y_val, y_val_pred, pos_label='Real', average='binary')
print(f"\nF1-Score برای کلاس Real: {f1_val:.4f}")

# محاسبه f1_score برای هر دو کلاس
f1_macro = f1_score(y_val, y_val_pred, average='macro')
print(f"F1-Score (macro average): {f1_macro:.4f}")

print("\nگزارش دسته‌بندی:")
print(classification_report(y_val, y_val_pred))

# 7. اگر نتایج رضایت‌بخش نبود، از روش دیگری استفاده می‌کنیم
if f1_macro < 0.80:
    print("\n=== تلاش با پارامترهای جدید ===")

    # کاهش تعداد مؤلفه‌های PCA برای جلوگیری از overfitting
    pca_new = PCA(n_components=500, random_state=42)
    X_train_pca_new = pca_new.fit_transform(X_train_scaled)
    X_test_pca_new = pca_new.transform(X_test_scaled)

    # تقسیم مجدد
    X_train_new, X_val_new, y_train_new, y_val_new = train_test_split(
        X_train_pca_new, y_train_full,
        test_size=0.2,
        random_state=42,
        stratify=y_train_full
    )

    # مدل جدید با kernel خطی که ممکن است برای داده‌های متنی بهتر عمل کند
    model_new = SVC(
        kernel='linear',
        C=1.0,
        random_state=42,
        class_weight='balanced'
    )

    model_new.fit(X_train_new, y_train_new)
    y_val_pred_new = model_new.predict(X_val_new)

    f1_macro_new = f1_score(y_val_new, y_val_pred_new, average='macro')
    print(f"F1-Score جدید (macro average): {f1_macro_new:.4f}")

    if f1_macro_new > f1_macro:
        model = model_new
        X_train_pca = X_train_pca_new
        X_test_pca = X_test_pca_new
        print("مدل جدید انتخاب شد!")

# 8. آموزش نهایی با کل داده‌های آموزشی
print("\n=== آموزش مدل نهایی ===")
model_final = SVC(
    kernel=model.kernel,
    C=model.C,
    gamma=model.gamma,
    random_state=42,
    class_weight='balanced'
)
model_final.fit(X_train_pca, y_train_full)
print("مدل نهایی آموزش داده شد!")

# 9. پیش‌بینی برای داده‌های تست
y_test_pred = model_final.predict(X_test_pca)

# 10. ساخت دیتافریم submission
submission = pd.DataFrame()
submission['label'] = y_test_pred

print("\n=== نتایج نهایی ===")
print(f"تعداد کل پیش‌بینی‌ها: {len(submission)}")
print(f"توزیع پیش‌بینی‌ها:")
print(submission['label'].value_counts())
print("\nنمونه‌ای از پیش‌بینی‌ها:")
print(submission.head(10))

# 11. ذخیره نتایج
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

joblib.dump(model_final, 'model')
submission.to_csv('submission.csv', index=False)
file_names = ['fake_news_detection.ipynb', 'submission.csv', 'model']
compress(file_names)
```

```python
import numpy as np
import pandas as pd
import scipy.sparse as ss

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, f1_score
from sklearn.preprocessing import StandardScaler

# خواندن داده‌ها
train_data = pd.read_csv('news_train.csv')
train_data_text_vectors = ss.load_npz('news_train_text_vectors.npz')

test_data = pd.read_csv('news_test.csv')
test_data_text_vectors = ss.load_npz('news_test_text_vectors.npz')

# تبدیل ماتریس sparse به dense
X_train_text = train_data_text_vectors.toarray()
X_test_text = test_data_text_vectors.toarray()

# استخراج برچسب‌ها
y_train = train_data['label'].values

print(f"تعداد نمونه‌های آموزشی: {len(train_data)}")
print(f"تعداد نمونه‌های تست: {len(test_data)}")

# =================
# مهندسی ویژگی جدید
# =================

def create_author_features(train_df, test_df, y_train):
    """محاسبه احتمال fake بودن بر اساس نویسنده"""
    # محاسبه نسبت fake برای هر نویسنده در داده آموزش
    author_fake_ratio = {}

    for author, label in zip(train_df['author'], y_train):
        if author not in author_fake_ratio:
            author_fake_ratio[author] = {'fake': 0, 'real': 0}

        if label == 'Fake':
            author_fake_ratio[author]['fake'] += 1
        else:
            author_fake_ratio[author]['real'] += 1

    # محاسبه نسبت
    for author in author_fake_ratio:
        total = author_fake_ratio[author]['fake'] + author_fake_ratio[author]['real']
        if total > 0:
            author_fake_ratio[author] = author_fake_ratio[author]['fake'] / total
        else:
            author_fake_ratio[author] = 0.5  # مقدار پیش‌فرض

    # اعمال به داده‌های train و test
    train_author_score = train_df['author'].map(lambda x: author_fake_ratio.get(x, 0.5))
    test_author_score = test_df['author'].map(lambda x: author_fake_ratio.get(x, 0.5))

    return train_author_score.values.reshape(-1, 1), test_author_score.values.reshape(-1, 1)

def create_site_features(train_df, test_df, y_train):
    """محاسبه احتمال fake بودن بر اساس سایت"""
    site_fake_ratio = {}

    for site, label in zip(train_df['site_url'], y_train):
        if site not in site_fake_ratio:
            site_fake_ratio[site] = {'fake': 0, 'real': 0}

        if label == 'Fake':
            site_fake_ratio[site]['fake'] += 1
        else:
            site_fake_ratio[site]['real'] += 1

    # محاسبه نسبت
    for site in site_fake_ratio:
        total = site_fake_ratio[site]['fake'] + site_fake_ratio[site]['real']
        if total > 0:
            site_fake_ratio[site] = site_fake_ratio[site]['fake'] / total
        else:
            site_fake_ratio[site] = 0.5

    train_site_score = train_df['site_url'].map(lambda x: site_fake_ratio.get(x, 0.5))
    test_site_score = test_df['site_url'].map(lambda x: site_fake_ratio.get(x, 0.5))

    return train_site_score.values.reshape(-1, 1), test_site_score.values.reshape(-1, 1)

def create_type_features(train_df, test_df, y_train):
    """محاسبه احتمال fake بودن بر اساس نوع"""
    type_fake_ratio = {}

    for news_type, label in zip(train_df['type'], y_train):
        if news_type not in type_fake_ratio:
            type_fake_ratio[news_type] = {'fake': 0, 'real': 0}

        if label == 'Fake':
            type_fake_ratio[news_type]['fake'] += 1
        else:
            type_fake_ratio[news_type]['real'] += 1

    # محاسبه نسبت
    for news_type in type_fake_ratio:
        total = type_fake_ratio[news_type]['fake'] + type_fake_ratio[news_type]['real']
        if total > 0:
            type_fake_ratio[news_type] = type_fake_ratio[news_type]['fake'] / total
        else:
            type_fake_ratio[news_type] = 0.5

    train_type_score = train_df['type'].map(lambda x: type_fake_ratio.get(x, 0.5))
    test_type_score = test_df['type'].map(lambda x: type_fake_ratio.get(x, 0.5))

    return train_type_score.values.reshape(-1, 1), test_type_score.values.reshape(-1, 1)

# ایجاد ویژگی‌های جدید
print("\nایجاد ویژگی‌های جدید...")
train_author_feat, test_author_feat = create_author_features(train_data, test_data, y_train)
train_site_feat, test_site_feat = create_site_features(train_data, test_data, y_train)
train_type_feat, test_type_feat = create_type_features(train_data, test_data, y_train)

# =================
# مرحله 1: آموزش SVM روی متن
# =================

print("\nمرحله 1: آموزش SVM روی ویژگی‌های متنی...")

# کاهش ابعاد با PCA
pca_text = PCA(n_components=500, random_state=42)
X_train_text_pca = pca_text.fit_transform(X_train_text)
X_test_text_pca = pca_text.transform(X_test_text)

print(f"ابعاد بعد از PCA: {X_train_text_pca.shape}")

# آموزش SVM روی متن
svm_text = SVC(kernel='rbf', C=10, gamma='scale',
               probability=True, random_state=42)
svm_text.fit(X_train_text_pca, y_train)

# پیش‌بینی احتمالات برای متن
train_text_proba = svm_text.predict_proba(X_train_text_pca)[:, 0].reshape(-1, 1)  # احتمال Fake
test_text_proba = svm_text.predict_proba(X_test_text_pca)[:, 0].reshape(-1, 1)

print("SVM متن آموزش داده شد!")

# =================
# مرحله 2: ترکیب همه ویژگی‌ها و آموزش SVM نهایی
# =================

print("\nمرحله 2: ترکیب ویژگی‌ها و آموزش مدل نهایی...")

# ترکیب همه ویژگی‌ها
X_train_combined = np.hstack([
    train_text_proba,
    train_author_feat,
    train_site_feat,
    train_type_feat
])

X_test_combined = np.hstack([
    test_text_proba,
    test_author_feat,
    test_site_feat,
    test_type_feat
])

print(f"تعداد ویژگی‌های نهایی: {X_train_combined.shape[1]}")

# نرمال‌سازی ویژگی‌های ترکیبی
scaler = StandardScaler()
X_train_combined_scaled = scaler.fit_transform(X_train_combined)
X_test_combined_scaled = scaler.transform(X_test_combined)

# جداسازی داده‌های اعتبارسنجی
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_combined_scaled, y_train, test_size=0.2, random_state=42, stratify=y_train
)

# آموزش مدل نهایی
model = SVC(kernel='rbf', C=100, gamma='scale', random_state=42)
model.fit(X_train_split, y_train_split)

# ارزیابی
y_val_pred = model.predict(X_val_split)
f1 = f1_score(y_val_split, y_val_pred, pos_label='Real', average='binary')
print(f"\nF1-Score روی داده‌های اعتبارسنجی: {f1:.4f}")

# گزارش کامل
print("\nگزارش دسته‌بندی:")
print(classification_report(y_val_split, y_val_pred))

# اگر نتیجه خوب بود، مدل را با کل داده آموزش بده
if f1 >= 0.80:
    print("\nآموزش مدل نهایی با کل داده‌ها...")
    model.fit(X_train_combined_scaled, y_train)
else:
    # تلاش با پارامترهای مختلف
    print("\nبهینه‌سازی پارامترها...")
    from sklearn.model_selection import GridSearchCV

    param_grid = {
        'C': [10, 50, 100, 200],
        'gamma': ['scale', 'auto', 0.001, 0.01, 0.1],
    }

    grid_search = GridSearchCV(
        SVC(kernel='rbf', random_state=42),
        param_grid,
        cv=5,
        scoring='f1_weighted',
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train_combined_scaled, y_train)
    model = grid_search.best_estimator_

    print(f"بهترین پارامترها: {grid_search.best_params_}")
    print(f"بهترین F1-Score: {grid_search.best_score_:.4f}")

# پیش‌بینی برای داده‌های تست
print("\nپیش‌بینی برای داده‌های تست...")
y_test_pred = model.predict(X_test_combined_scaled)

# ایجاد DataFrame برای submission
submission = pd.DataFrame({
    'label': y_test_pred
})

print(f"\nتعداد کل پیش‌بینی‌ها: {len(submission)}")
print(f"توزیع پیش‌بینی‌ها:\n{submission['label'].value_counts()}")
print("\nنمونه‌ای از پیش‌بینی‌ها:")
print(submission.head(10))

# ذخیره نتایج
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

joblib.dump(model, 'model')
submission.to_csv('submission.csv', index=False)
file_names = ['fake_news_detection.ipynb', 'submission.csv', 'model']
compress(file_names)
```
