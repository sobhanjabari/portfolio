</h2>

<ol dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <li><b>نرمال‌سازی (Normalization):</b> استانداردسازی حروف و کلمات (مانند تبدیل "ي" به "ی" و "ك" به "ک")</li>
  <li><b>حذف علائم نگارشی:</b> حذف کاراکترهای غیرمتنی (مانند ؟!،%،$ و ...) که در پردازش زبانی نقشی ندارند.</li>
  <li><b>حذف اعداد:</b> معمولاً اعداد در تحلیل‌های زبانی بی‌اهمیت هستند، مگر در کاربردهای خاص.</li>
  <li><b>توکن‌سازی (Tokenization):</b> شکستن متن به کلمات یا واحدهای معنایی.</li>
  <li><b>ریشه‌یابی (Lemmatization):</b> تبدیل کلمات به شکل پایه‌شان. مثلاً «رفتیم» به «رفتن» تبدیل می‌شود.</li>

</ol>

```python
lemmatizer = Lemmatizer()
normalizer = Normalizer()
tokenizer = WordTokenizer()
stop_words = set(stopwords_list())

def persian_text_preprocessing(text):
    # ۱) نرمال‌سازی
    text = normalizer.normalize(text)

    # ۲) حذف علائم نگارشی و اعداد
    text = re.sub(r'[^\w\s]', ' ', text)  # حذف علائم
    text = re.sub(r'\d+', ' ', text)      # حذف اعداد

    # ۳) توکن‌سازی
    tokens = tokenizer.tokenize(text)

    # ۴) حذف کلمات توقف و ریشه‌یابی
    cleaned_tokens = []
    for token in tokens:
        if token not in stop_words and len(token) > 1:
            lemma = lemmatizer.lemmatize(token)
            cleaned_tokens.append(lemma)

    return " ".join(cleaned_tokens)

# اعمال پیش‌پردازش روی ستون content
data["content"] = data["content"].apply(persian_text_preprocessing)
```

```python
def recommend_news(title, num_recommendations=5):
    # پیدا کردن ایندکس خبر با عنوان داده‌شده
    idx = data[data['title'] == title].index[0]

    # بدست آوردن لیست شباهت‌ها با همه خبرها
    sim_scores = list(enumerate(cosine_sim[idx]))

    # مرتب‌سازی بر اساس بیشترین شباهت، حذف خود خبر
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]

    # گرفتن ایندکس‌ها
    news_indices = [i[0] for i in sim_scores]

    return data['title'].iloc[news_indices]

# تست پیشنهادها
recommend_news(data['title'].iloc[0], 5)
```

```python
# ---------------------------------
# گام 1 : فراخوانی کتابخانه‌ها
# ---------------------------------
!pip install hazm

import numpy as np
import pandas as pd
import re
from hazm import Normalizer, WordTokenizer, Lemmatizer, stopwords_list
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("Numpy version:", np.__version__)

# ---------------------------------
# گام 2 : خواندن فایل داده‌ها
# ---------------------------------
df = pd.read_csv("news_data.csv")
print(df.columns)
df.head()

# ---------------------------------
# گام 3 : بررسی اولیه داده‌ها
# ---------------------------------
print(df.info())
print(df.sample(5))

# ---------------------------------
# گام 4 : پیش‌پردازش متن
# ---------------------------------
data = df.copy()

lemmatizer = Lemmatizer()
normalizer = Normalizer()
tokenizer = WordTokenizer()
stop_words = set(stopwords_list())

def persian_text_preprocessing(text):
    text = str(text)
    text = normalizer.normalize(text)
    text = re.sub(r'[^\w\s]', ' ', text)  # حذف علائم
    text = re.sub(r'\d+', ' ', text)      # حذف اعداد
    tokens = tokenizer.tokenize(text)
    cleaned_tokens = [
        lemmatizer.lemmatize(tok)
        for tok in tokens
        if tok not in stop_words and len(tok) > 1
    ]
    return " ".join(cleaned_tokens)

data["content"] = data["content"].apply(persian_text_preprocessing)

# ---------------------------------
# گام 5 : ایجاد ستون tag
# ---------------------------------
data['tag'] = data['label'] + " " + data['content']

# ---------------------------------
# گام 6 : بردارسازی با TF-IDF
# ---------------------------------
tfidf = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf.fit_transform(data['tag'])
print("TF-IDF matrix shape:", tfidf_matrix.shape)

# ---------------------------------
# گام 7 : محاسبه شباهت کسینوسی
# ---------------------------------
cosine_sim = cosine_similarity(tfidf_matrix)
print("Cosine similarity shape:", cosine_sim.shape)

# ---------------------------------
# گام 8 : تابع پیشنهاددهنده بر اساس index
# ---------------------------------
def recommend_by_index(idx, num_recommendations=5):
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]  # حذف خودش
    indices = [i[0] for i in sim_scores]
    return indices

# ---------------------------------
# گام 9 : تست سیستم روی خبر شماره 46
# ---------------------------------
base_index = 46
recommended_indices = recommend_by_index(base_index, 5)

print("خبر پایه:\n", data.iloc[base_index]['content'])
print("\nپیشنهادها:")
for i in recommended_indices:
    print(i, "-", data.iloc[i]['label'])

# ---------------------------------
# گام 10 : ماموریت نهایی - ساخت submission
# ---------------------------------
submission = pd.DataFrame({
    "label": data.iloc[recommended_indices]['label'].values,
    "index": recommended_indices
})

print("\nنتیجه submission:")
print(submission)

# ذخیره در فایل CSV
submission.to_csv("submission.csv", index=False)
print("\n📁 فایل submission.csv ذخیره شد.")
```
