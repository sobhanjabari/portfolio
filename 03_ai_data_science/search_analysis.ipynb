<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
در ابتدا قصد داریم درصد محبوبیت هرکدام از سرویس‌ها (اتوبوس، هواپیما، تاکسی و غیره) را بررسی کنیم. برای این کار نیاز است به ازای هر سرویس (ستون <code>ServiceType</code>) محاسبه کنید که چند درصد از کل جست‌و‌جوها را شامل می‌شود. پاسخ خود را به شکل یک دیکشنری و همچون فرمت زیر در متغیر <code>percentages</code> ذخیره کنید. سپس به کمک کدی که برای‌تان آماده شده می‌توانید نمودار دایره‌ای (Pie Chart) آن را رسم کنید.
</font>

```python
# محاسبه درصد محبوبیت هر سرویس
service_counts = data['ServiceType'].value_counts(normalize=True)

percentages = {
    'bus': service_counts.get('bus', 0),
    'flight': service_counts.get('flight', 0),
    'train': service_counts.get('train', 0),
    'hotel': service_counts.get('hotel', 0),
    'taxi': service_counts.get('taxi', 0),
    'ship': service_counts.get('ship', 0)
}

print("درصد محبوبیت سرویس‌ها:")
for service, percentage in percentages.items():
    print(f"{service}: {percentage:.4f}")

# رسم نمودار دایره‌ای
import plotly.express as px
fig = px.pie(values=list(percentages.values()),
             names=list(percentages.keys()),
             title='Pie Chart of Service Types')
fig.show()
```

```python
preprocessed_data = data.copy()

# پیدا کردن ردیف‌هایی که شامل '- پایانه' هستند
mask = preprocessed_data['AcceptString'].str.contains('- پایانه', na=False)

# اصلاح: استفاده از روش صحیح برای جدا کردن نام شهر
preprocessed_data.loc[mask, 'AcceptString'] = (
    preprocessed_data.loc[mask, 'AcceptString']
    .apply(lambda x: x[:x.find(' -')] if ' -' in x else x)
)

print(f"تعداد ردیف‌های تغییر یافته: {mask.sum()}")
preprocessed_data.head()
```

```python
# فیلتر شهرهای با جمعیت بالای 500 هزار نفر
large_cities_mask = cities['2016 Census'] > 500000
large_cities = cities[large_cities_mask]['City FA'].tolist()

print(f"تعداد شهرهای با جمعیت بالای 500 هزار: {len(large_cities)}")
print("شهرهای پرجمعیت:")
for city in large_cities[:10]:  # نمایش 10 مورد اول
    population = cities[cities['City FA'] == city]['2016 Census'].iloc[0]
    print(f" {city}: {population:,.0f} نفر")

# یافتن شهرهایی که در لیست top_cities نیستند
not_in_top = [city for city in large_cities if city not in top_cities]
print(f"\nشهرهای پرجمعیت که در top 20 نیستند ({len(not_in_top)} شهر):")
for city in not_in_top:
    population = cities[cities['City FA'] == city]['2016 Census'].iloc[0]
    print(f" {city}: {population:,.0f} نفر")
```

```python
import zipfile
import joblib
import json
import os

if not os.path.exists(os.path.join(os.getcwd(), 'search_analysis.ipynb')):
    %notebook -e text_categorization.ipynb


def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

json.dump(percentages, open('percentages.json', 'w'))
preprocessed_data.to_csv('preprocessed_data.csv', index=False)
data_transport.to_csv('data_transport.csv', index=False)
json.dump(top_cities, open('top_cities.json', 'w'))
json.dump(top_provinces, open('top_provinces.json', 'w'))
json.dump(not_in_top, open('not_in_top.json', 'w'))

file_names = ['search_analysis.ipynb', 'percentages.json', 'preprocessed_data.csv',
              'data_transport.csv', 'top_cities.json', 'top_provinces.json', 'not_in_top.json']
compress(file_names)
```
