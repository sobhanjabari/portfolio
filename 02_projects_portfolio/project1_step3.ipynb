```python
# گروه‌بندی بر اساس Company و Review Date و محاسبه میانگین Rating
companies = df.groupby(['Company', 'Review Date'])['Rating'].mean().unstack()

# حذف نام‌های ستون‌ها و ایندکس
companies.columns.name = None
companies.index.name = None

companies.head()
```

```python
# انتخاب ستون‌های سال 2012 به بعد
years_2012_after = [col for col in companies.columns if col >= 2012]

# محاسبه میانگین امتیاز برای هر شرکت
mean_ratings = companies[years_2012_after].mean(axis=1)

# حذف نام سری و ایندکس
mean_ratings.name = None
mean_ratings.index.name = None

print(f"تعداد شرکت‌ها: {len(mean_ratings)}")
print(mean_ratings.head())
```

```python
# کپی کردن mean_ratings
temp_df = mean_ratings.copy().reset_index()

# مرتب‌سازی بر اساس میانگین امتیاز (نزولی) و نام شرکت (صعودی)
temp_df = temp_df.sort_values([0, 'index'], ascending=[False, True])

# انتخاب 10 شرکت برتر
top_10_companies = temp_df.head(10)

# تبدیل به سری با ایندکس شرکت‌ها
best_ratings = top_10_companies.set_index('index')[0]

# حذف نام‌های سری و ایندکس
best_ratings.name = None
best_ratings.index.name = None

print("10 شرکت برتر:")
print(best_ratings)
```

```python
# انتخاب شکلات‌هایی که شرکت‌سازشان جزو 10 شرکت برتر است
chocolates_to_sell = df[df['Company'].isin(best_ratings.index)]

# ریست کردن ایندکس
chocolates_to_sell.reset_index(drop=True, inplace=True)

print(f"تعداد شکلات‌های انتخاب شده: {len(chocolates_to_sell)}")
chocolates_to_sell.head()
```

```python
import zlib
import zipfile

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

df.to_csv('normal_chocolates.csv', index = True)
companies.to_csv('companies.csv', index = True)
chocolates_to_sell.to_csv('chocolates_to_sell.csv', index = True)
mean_ratings.to_csv("mean_ratings.csv", index = True)
best_ratings.to_csv("best_ratings.csv", index = True)

np.savez("answers.npz" ,priceSum = priceSum)
file_names = ["answers.npz", "project1_step3.ipynb", "chocolates_to_sell.csv",
              "companies.csv", "mean_ratings.csv", "best_ratings.csv", "normal_chocolates.csv"]
compress(file_names)
```
