```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# خواندن داده‌ها
df = pd.read_csv('preprocessed_sales.csv')

# حذف United Kingdom و محاسبه TotalPrice
df_filtered = df[df['Country'] != 'United Kingdom'].copy()
df_filtered['TotalPrice'] = df_filtered['UnitPrice'] * df_filtered['Quantity']

# گروه‌بندی و محاسبه آمار
country_stats = df_filtered.groupby('Country').agg({
    'CustomerId': 'nunique',
    'TotalPrice': 'sum'
}).rename(columns={
    'CustomerId': 'customer_count',
    'TotalPrice': 'sales_amount'
})

# محاسبه چارک سوم
q_sales = country_stats['sales_amount'].quantile(0.75)
q_customers = country_stats['customer_count'].quantile(0.75)

# تعریف گروه‌ها
def assign_group(row):
    SA = row['sales_amount']
    CC = row['customer_count']

    if SA > q_sales and CC > q_customers:
        return 'Highest customer & revenue'
    elif SA < q_sales and CC > q_customers:
        return 'High customer & low revenue'
    elif SA > q_sales and CC < q_customers:
        return 'High revenue & low customer'
    else:
        return 'Low customer & revenue'

country_stats['group'] = country_stats.apply(assign_group, axis=1)

# رنگ‌ها
colors = {
    'Highest customer & revenue': 'red',
    'High customer & low revenue': 'blue',
    'High revenue & low customer': 'green',
    'Low customer & revenue': 'cyan'
}

# رسم نمودار
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(10, 10))

# رسم گروه‌ها
groups = ['Highest customer & revenue', 'High customer & low revenue',
          'High revenue & low customer', 'Low customer & revenue']

for group in groups:
    group_data = country_stats[country_stats['group'] == group]
    ax.scatter(group_data['sales_amount'],
              group_data['customer_count'],
              c=colors[group],
              s=20,
              label=group)

# اضافه کردن نام کشورها
countries_to_label = ['France', 'Spain', 'Netherlands']
for country in countries_to_label:
    if country in country_stats.index:
        x = country_stats.loc[country, 'sales_amount']
        y = country_stats.loc[country, 'customer_count']
        ax.text(x, y + 0.5, country, fontsize=15, ha='center')

# تنظیمات نمودار
ax.set_title('Market Study', fontsize=18)
ax.set_xlabel('Sales amount', fontsize=16)
ax.set_ylabel('Number of customers', fontsize=16)

# Legend
ax.legend(title='Group',
         frameon=True,
         facecolor='silver',
         title_fontsize=16,
         fontsize=14)

plt.tight_layout()
```
