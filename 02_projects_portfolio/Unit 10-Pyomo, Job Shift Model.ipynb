```python
import pyomo.environ as pyo
m1 = pyo.ConcreteModel()
```

```python
m1.x = pyo.Var((1,2,3,4,5,6,7), domain = pyo.NonNegativeIntegers)
m1.x.pprint()
```

```python
m1.c1 = pyo.Constraint(expr = m1.x[1] + m1.x[4] + m1.x[5] + m1.x[6] + m1.x[7] >= 17)
m1.c2 = pyo.Constraint(expr = m1.x[1] + m1.x[2] + m1.x[5] + m1.x[6] + m1.x[7] >= 13)
m1.c3 = pyo.Constraint(expr = m1.x[1] + m1.x[2] + m1.x[3] + m1.x[6] + m1.x[7] >= 15)
m1.c4 = pyo.Constraint(expr = m1.x[1] + m1.x[2] + m1.x[3] + m1.x[4] + m1.x[7] >= 19)
m1.c5 = pyo.Constraint(expr = m1.x[1] + m1.x[2] + m1.x[3] + m1.x[4] + m1.x[5] >= 14)
m1.c6 = pyo.Constraint(expr = m1.x[2] + m1.x[3] + m1.x[4] + m1.x[5] + m1.x[6] >= 16)
m1.c7 = pyo.Constraint(expr = m1.x[3] + m1.x[4] + m1.x[5] + m1.x[6] + m1.x[7] >= 11)
```

```python
m1.obj = pyo.Objective(expr = sum(m1.x[i] for i in range (1,8)) , sense=pyo.minimize)
```

```python
m1.pprint()
```

```python
solver_name = 'cplex'
solver = pyo.SolverFactory(solver_name)
result = solver.solve(m1)
print(result.solver.Termination_condition)
```

```python
import pyomo.environ as pyo

# Create the model
m3 = pyo.ConcreteModel()

# Define days of the week
week_days = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Correctly compute shift patterns (workers starting on these days cover the target day)
shift_patterns = {
    day: [
        week_days[(i - offset) % 7]  # Cyclic 5-day shifts starting from (day -4) to day
        for offset in range(4, -1, -1)
    ]
    for i, day in enumerate(week_days)
}

# Workforce demand per day
workforce = {
    'Saturday': 17,
    'Sunday': 13,
    'Monday': 15,
    'Tuesday': 19,
    'Wednesday': 14,
    'Thursday': 16,
    'Friday': 11
}

# Define decision variables
m3.x = pyo.Var(week_days, domain=pyo.NonNegativeIntegers)

# Constraint rule: Ensure coverage for each day
def constraint_rule(model, day):
    return sum(model.x[start_day] for start_day in shift_patterns[day]) >= workforce[day]

# Add constraints
m3.coverage = pyo.Constraint(week_days, rule=constraint_rule)

# Objective: Minimize total workers
m3.obj = pyo.Objective(expr=sum(m3.x[day] for day in week_days), sense=pyo.minimize)

# Solve with CPLEX
solver = pyo.SolverFactory('cplex')
result = solver.solve(m3)

# Print results
print(f"Solver Status: {result.solver.Termination_condition}\n")
for day in week_days:
    print(f"Workers starting on {day}: {pyo.value(m3.x[day])}")
print("-" * 40)
print(f"Total Workers: {pyo.value(m3.obj)}")
```

```python
!pip install highspy glpk
import pyomo.environ as pyo
import pandas as pd
import numpy as np
import sys

# ----------------------------
# 1st Format (Without Advertising)
# ----------------------------
def build_basic_blending_model(solver_name='appsi_highs'):
    """Build and solve the basic blending model without advertising"""
    model = pyo.ConcreteModel()

    # Sets
    model.gas = pyo.Set(initialize=['gas1', 'gas2', 'gas3'])
    model.oil = pyo.Set(initialize=['oil1', 'oil2', 'oil3'])

    # Parameters
    profit_data = {'gas1': 70, 'gas2': 60, 'gas3': 50}
    cost_data = {'oil1': 45, 'oil2': 35, 'oil3': 25}
    demand_data = {'gas1': 3000, 'gas2': 2000, 'gas3': 1000}
    purchase_data = {'oil1': 5000, 'oil2': 5000, 'oil3': 5000}
    oktan_data = {'oil1': 12, 'oil2': 6, 'oil3': 8}
    solfor_data = {'oil1': 0.005, 'oil2': 0.02, 'oil3': 0.03}
    quality_needs = {
        'oktan': {'gas1': 10, 'gas2': 8, 'gas3': 6},
        'solfor': {'gas1': 0.01, 'gas2': 0.02, 'gas3': 0.01}
    }

    # Initialize parameters
    model.gas_profit = pyo.Param(model.gas, initialize=profit_data)
    model.oil_cost = pyo.Param(model.oil, initialize=cost_data)
    model.demands = pyo.Param(model.gas, initialize=demand_data)
    model.oil_purchase = pyo.Param(model.oil, initialize=purchase_data)
    model.oktan_portion = pyo.Param(model.oil, initialize=oktan_data)
    model.solfor_portion = pyo.Param(model.oil, initialize=solfor_data)
    model.oktan_need = pyo.Param(model.gas, initialize=quality_needs['oktan'])
    model.solfor_need = pyo.Param(model.gas, initialize=quality_needs['solfor'])

    # Variables
    model.x = pyo.Var(model.oil, model.gas, domain=pyo.NonNegativeReals)

    # Objective
    def obj_rule(m):
        income = sum(m.gas_profit[g] * m.x[o,g] for o in m.oil for g in m.gas)
        cost = sum(m.oil_cost[o] * m.x[o,g] for o in m.oil for g in m.gas)
        fixed_cost = 4 * sum(m.x[o,g] for o in m.oil for g in m.gas)
        return income - cost - fixed_cost
    model.obj = pyo.Objective(rule=obj_rule, sense=pyo.maximize)

    # Constraints
    def demand_constraint(m, g):
        return sum(m.x[o,g] for o in m.oil) == m.demands[g]
    model.demand_constr = pyo.Constraint(model.gas, rule=demand_constraint)

    def purchase_constraint(m, o):
        return sum(m.x[o,g] for g in m.gas) <= m.oil_purchase[o]
    model.purchase_constr = pyo.Constraint(model.oil, rule=purchase_constraint)

    def total_capacity(m):
        return sum(m.x[o,g] for o in m.oil for g in m.gas) <= 14000
    model.capacity_constr = pyo.Constraint(rule=total_capacity)

    def oktan_constraint(m, g):
        return sum(m.oktan_portion[o] * m.x[o,g] for o in m.oil) >= \
               m.oktan_need[g] * sum(m.x[o,g] for o in m.oil)
    model.oktan_constr = pyo.Constraint(model.gas, rule=oktan_constraint)

    def solfor_constraint(m, g):
        return sum(m.solfor_portion[o] * m.x[o,g] for o in m.oil) <= \
               m.solfor_need[g] * sum(m.x[o,g] for o in m.oil)
    model.solfor_constr = pyo.Constraint(model.gas, rule=solfor_constraint)

    # Solve with automatic solver selection
    solve_model(model, solver_name)
    return model

# ----------------------------
# 2nd Format (With Advertising)
# ----------------------------
def build_advertising_model(solver_name='appsi_highs'):
    """Build and solve model with advertising decision variables"""
    model = pyo.ConcreteModel()

    # Inherit all parameters from basic model
    build_basic_blending_model().transfer_components_to(model)

    # Add advertising variables
    model.y = pyo.Var(model.gas, domain=pyo.NonNegativeReals)

    # Modify objective function
    model.obj.deactivate()
    def advertising_obj(m):
        base_profit = sum(m.gas_profit[g] * m.x[o,g] for o in m.oil for g in m.gas)
        costs = sum(m.oil_cost[o] * m.x[o,g] for o in m.oil for g in m.gas)
        fixed_cost = 4 * sum(m.x[o,g] for o in m.oil for g in m.gas)
        ad_cost = sum(m.y[g] for g in m.gas)
        return base_profit - costs - fixed_cost - ad_cost
    model.obj = pyo.Objective(rule=advertising_obj, sense=pyo.maximize)

    # Modify demand constraints
    model.demand_constr.deactivate()
    def ad_demand_constraint(m, g):
        return sum(m.x[o,g] for o in m.oil) == m.demands[g] + 10*m.y[g]
    model.ad_demand_constr = pyo.Constraint(model.gas, rule=ad_demand_constraint)

    # Solve and return
    solve_model(model, solver_name)
    return model

# ----------------------------
# Solver Handling Utilities
# ----------------------------
def solve_model(model, preferred_solver):
    """Generic solver handler with fallback logic"""
    solvers = [preferred_solver, 'glpk', 'appsi_highs', 'cbc']

    for solver_name in solvers:
        solver = pyo.SolverFactory(solver_name)
        if solver.available():
            result = solver.solve(model)
            if str(result.solver.termination_condition) == "optimal":
                print(f"Solved successfully with {solver_name}")
                return
            else:
                print(f"Solver {solver_name} failed with status: {result.solver.termination_condition}")
        else:
            print(f"Solver {solver_name} not available")

    raise RuntimeError("No available solver could solve the model")

# ----------------------------
# Data Import/Export
# ----------------------------
def excel_interface(input_file='blendingparams.xlsx'):
    """Build model from Excel data"""
    try:
        gas_params = pd.read_excel(input_file, sheet_name='gas', index_col=0)
        oil_params = pd.read_excel(input_file, sheet_name='oil', index_col=0)
    except FileNotFoundError:
        print(f"Error: Input file {input_file} not found!")
        sys.exit(1)

    model = pyo.ConcreteModel()

    # Create sets
    model.gas = pyo.Set(initialize=gas_params.index.tolist())
    model.oil = pyo.Set(initialize=oil_params.index.tolist())

    # Load parameters
    param_mapping = {
        'gas': {
            'demand': 'demands',
            'profit': 'gas_profit',
            'oktan_need': 'oktan_need',
            'solfor_need': 'solfor_need'
        },
        'oil': {
            'cost': 'oil_cost',
            'purchase': 'oil_purchase',
            'oktan_portion': 'oktan_portion',
            'solfor_portion': 'solfor_portion'
        }
    }

    for sheet, params in param_mapping.items():
        df = gas_params if sheet == 'gas' else oil_params
        for col, param_name in params.items():
            data = df[col].to_dict()
            setattr(model, param_name, pyo.Param(getattr(model, sheet), initialize=data))

    # Add variables and constraints (similar to previous models)
    # ... [rest of model setup] ...

    return model

def export_results(model, output_file='results.xlsx'):
    """Export model results to Excel"""
    results = []
    for o in model.oil:
        for g in model.gas:
            results.append({
                'Oil': o,
                'Gas': g,
                'Amount': pyo.value(model.x[o,g]),
                'Cost': pyo.value(model.oil_cost[o]) * pyo.value(model.x[o,g])
            })

    df = pd.DataFrame(results)
    try:
        df.to_excel(output_file, index=False)
        print(f"Results exported to {output_file}")
    except PermissionError:
        print(f"Error: Could not write to {output_file} - file may be open!")

# ----------------------------
# Execution
# ----------------------------
if __name__ == "__main__":
    # Basic model
    print("Solving basic blending model:")
    basic_model = build_basic_blending_model()
    export_results(basic_model, 'basic_results.xlsx')

    # Advertising model
    print("\nSolving advertising model:")
    ad_model = build_advertising_model()
    export_results(ad_model, 'ad_results.xlsx')

    # Excel interface
    print("\nBuilding model from Excel:")
    excel_model = excel_interface()
```

```python
# Install required packages (run once)
# !pip install highspy pandas pyomo
!pip install highspy pandas pyomo
import pyomo.environ as pyo  # Pyomo optimization modeling
import pandas as pd  # Data handling

# ====================== MODEL BUILDER ======================
def build_blending_model(advertising=False):
    """Construct Pyomo model for gasoline blending optimization

    Args:
        advertising (bool): Whether to include advertising decision variables

    Returns:
        pyo.ConcreteModel: Configured optimization model
    """

    # Initialize concrete model
    model = pyo.ConcreteModel()

    # ----- SETS -----
    # Define fuel types and crude oil sources
    model.gas = pyo.Set(initialize=['gas1', 'gas2', 'gas3'])  # Gasoline types
    model.oil = pyo.Set(initialize=['oil1', 'oil2', 'oil3'])   # Crude oil types

    # ----- PARAMETERS -----
    # Define all model parameters using dictionary
    params = {
        # Gasoline selling prices ($/unit)
        'gas_profit': {'gas1':70, 'gas2':60, 'gas3':50},
        # Crude oil purchase costs ($/unit)
        'oil_cost': {'oil1':45, 'oil2':35, 'oil3':25},
        # Base market demands (units)
        'demands': {'gas1':3000, 'gas2':2000, 'gas3':1000},
        # Maximum crude oil availability (units)
        'oil_purchase': {'oil1':5000, 'oil2':5000, 'oil3':5000},
        # Octane ratings (oil) and requirements (gas)
        'oktan': {'oil1':12, 'oil2':6, 'oil3':8, 'gas1':10, 'gas2':8, 'gas3':6},
        # Sulfur content (oil) and limits (gas)
        'solfor': {'oil1':0.005, 'oil2':0.02, 'oil3':0.03, 'gas1':0.01, 'gas2':0.02, 'gas3':0.01}
    }

    # Add parameters to model with automatic type detection
    for name, data in params.items():
        if 'gas' in name:
            # Gasoline-related parameters
            model.add_component(name, pyo.Param(model.gas, initialize=data))
        elif 'oil' in name:
            # Oil-related parameters
            model.add_component(name, pyo.Param(model.oil, initialize=data))
        else:
            # Mixed domain parameters (both oil and gas)
            model.add_component(name, pyo.Param(model.gas | model.oil, initialize=data))

    # ----- DECISION VARIABLES -----
    # Blending quantities (oil, gas)
    model.x = pyo.Var(model.oil, model.gas,
                    domain=pyo.NonNegativeReals,
                    doc='Amount of oil used in gasoline production')

    # Advertising budget variables (if enabled)
    if advertising:
        model.y = pyo.Var(model.gas,
                        domain=pyo.NonNegativeReals,
                        doc='Advertising budget for gasoline type')

    # ----- OBJECTIVE FUNCTION -----
    def obj_rule(m):
        """Calculate total profit: income - costs - advertising"""
        # Revenue from gasoline sales
        income = sum(m.gas_profit[g] * m.x[o,g] for o,g in m.oil*m.gas)

        # Production costs (oil + fixed)
        costs = sum(m.oil_cost[o] * m.x[o,g] + 4 * m.x[o,g]  # $4/unit processing
                  for o,g in m.oil*m.gas)

        # Add advertising costs if enabled
        if advertising:
            costs += sum(m.y[g] for g in m.gas)

        return income - costs

    model.obj = pyo.Objective(rule=obj_rule,
                            sense=pyo.maximize,
                            doc='Maximize total profit')

    # ----- CONSTRAINTS -----
    def demand_rule(m, g):
        """Meet base demand plus advertising-boosted demand"""
        rhs = m.demands[g] + (10 * m.y[g] if advertising else 0)
        return sum(m.x[o,g] for o in m.oil) == rhs

    model.demand = pyo.Constraint(model.gas,
                                rule=demand_rule,
                                doc='Demand fulfillment')

    # Crude oil purchase limits
    model.purchase_limit = pyo.Constraint(
        model.oil,
        rule=lambda m, o: sum(m.x[o,g] for g in m.gas) <= m.oil_purchase[o],
        doc='Oil purchase capacity'
    )

    # Quality constraints (octane and sulfur)
    for qual in ['oktan', 'solfor']:
        # Create constraint name
        constr_name = f'{qual}_quality'

        # Choose comparison operator based on quality type
        if qual == 'oktan':
            compare = lambda a, b: a >= b  # Octane minimum requirement
        else:
            compare = lambda a, b: a <= b  # Sulfur maximum limit

        # Create constraint rule
        def quality_rule(m, g, q=qual):
            total_quality = sum(getattr(m, q)[o] * m.x[o,g] for o in m.oil)
            total_volume = sum(m.x[o,g] for o in m.oil)
            return compare(total_quality, getattr(m, q)[g] * total_volume)

        # Add constraint to model
        model.add_component(constr_name,
                          pyo.Constraint(model.gas, rule=quality_rule))

    return model

# ====================== SOLVER HANDLER ======================
def solve_model(model, preferred_solver='appsi_highs'):
    """Solve model with solver fallback mechanism

    Args:
        model (pyo.ConcreteModel): Model to solve
        preferred_solver (str): Primary solver to try

    Returns:
        pyo.ConcreteModel: Solved model

    Raises:
        RuntimeError: If no available solver found
    """
    # Try solvers in priority order
    for solver_name in [preferred_solver, 'cbc', 'glpk']:
        solver = pyo.SolverFactory(solver_name)
        if solver.available():
            results = solver.solve(model)
            if str(results.solver.termination_condition) == 'optimal':
                print(f"Solved successfully with {solver_name}")
                return model
            else:
                print(f"Solver {solver_name} failed: {results.solver.termination_condition}")
        else:
            print(f"Solver {solver_name} not available")

    raise RuntimeError("No available solver could solve the model")

# ====================== MAIN EXECUTION ======================
if __name__ == "__main__":
    # ------ BASIC MODEL ------
    print("Building basic blending model...")
    basic_model = build_blending_model()
    solve_model(basic_model)

    # ------ ADVERTISING MODEL ------
    print("\nBuilding advertising model...")
    ad_model = build_blending_model(advertising=True)
    solve_model(ad_model)

    # ------ DATA EXPORT ------
    try:
        # Create DataFrame with blending results
        results_data = [
            (oil, gas, pyo.value(basic_model.x[oil, gas]))
            for oil in basic_model.oil
            for gas in basic_model.gas
        ]

        pd.DataFrame(results_data,
                   columns=['Crude Oil', 'Gasoline', 'Quantity']
                   ).to_excel('blending_results.xlsx', index=False)

        print("Results saved to blending_results.xlsx")

    except PermissionError:
        print("Error: Could not save results - file may be open in Excel")
```
