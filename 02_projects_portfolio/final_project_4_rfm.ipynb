<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
    اکنون نیاز است هرکدام از این سه معیار را براساس چارک اول،‌ دوم و سوم به شکل زیر به ۴ گروه مختلف تقسیم کنید
    (برای یادآوری مفهوم چارک،
    <a href="https://fa.wikipedia.org/wiki/%DA%86%D8%A7%D8%B1%DA%A9_(%D8%A2%D9%85%D8%A7%D8%B1)">این مطلب</a>

<center>
<div dir=rtl style="direction: rtl;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>

| <b>شماره گروه</b> | <b>شرط</b> |
| :---: | :---: |
| 1 | <code dir=ltr>value <= Q1</code> |
| 2 | <code dir=ltr>Q1 < value <= Q2</code> |
| 3 | <code dir=ltr>Q2 < value <= Q3</code> |
| 4 | <code dir=ltr>Q3 < value</code> |

</font>
</div>
</center>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
    به ازای هر کدام از این سه معیار چنین کاری را انجام داده و نتایج را در سه ستون دیگر به نام‌های
    <code>F_quartile</code>، <code>R_quartile</code> و <code>M_quartile</code>
    ذخیره کنید.
    توجه کنید که مقادیر این ستون‌ها معادل شماره گروه یعنی اعداد <code>1</code> تا <code>4</code> خواهد بود.
</font>
</p>

```python
df_customer_segments['R_quartile'] = None # To-Do
df_customer_segments['F_quartile'] = None # To-Do
df_customer_segments['M_quartile'] = None # To-Do
df_customer_segments.head()
```

```python
# محاسبه چارک‌ها برای Recency (توجه: برای Recency کمتر بهتر است)
r_labels = [1, 2, 3, 4]
df_customer_segments['R_quartile'] = pd.qcut(df_customer_segments['Recency'],
                                             q=4,
                                             labels=r_labels)

# محاسبه چارک‌ها برای Frequency
f_labels = [1, 2, 3, 4]
df_customer_segments['F_quartile'] = pd.qcut(df_customer_segments['Frequency'].rank(method='first'),
                                             q=4,
                                             labels=f_labels)

# محاسبه چارک‌ها برای MonetaryValue
m_labels = [1, 2, 3, 4]
df_customer_segments['M_quartile'] = pd.qcut(df_customer_segments['MonetaryValue'].rank(method='first'),
                                             q=4,
                                             labels=m_labels)

df_customer_segments.head()
```

```python
# تابعی برای محاسبه چارک‌ها و تخصیص گروه
def assign_quartile_groups(df, column):
    # محاسبه چارک‌ها
    q1 = df[column].quantile(0.25)
    q2 = df[column].quantile(0.50)
    q3 = df[column].quantile(0.75)

    # تخصیص گروه بر اساس شرایط جدول
    conditions = [
        df[column] <= q1,
        (df[column] > q1) & (df[column] <= q2),
        (df[column] > q2) & (df[column] <= q3),
        df[column] > q3
    ]
    choices = [1, 2, 3, 4]

    return np.select(conditions, choices)

# اعمال گروه‌بندی برای هر معیار
df_customer_segments['R_quartile'] = assign_quartile_groups(df_customer_segments, 'Recency')
df_customer_segments['F_quartile'] = assign_quartile_groups(df_customer_segments, 'Frequency')
df_customer_segments['M_quartile'] = assign_quartile_groups(df_customer_segments, 'MonetaryValue')

# تبدیل به integer
df_customer_segments['R_quartile'] = df_customer_segments['R_quartile'].astype(int)
df_customer_segments['F_quartile'] = df_customer_segments['F_quartile'].astype(int)
df_customer_segments['M_quartile'] = df_customer_segments['M_quartile'].astype(int)

df_customer_segments.head()
```

    گروه‌بندی مشتریان

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
    حال باید براساس مقادیر ستون <code>RFM</code> و طبق جدول زیر به گروه‌بندی مشتریان بپردازید و نام گروه آن‌ها را در ستونی به نام <code>Segment</code> ذخیره کنید.
    توجه داشته باشید که در این جدول، <code>X</code> به معنی هر مقداری است. به‌عنوان مثال <code>14X</code> به معنی <code>141</code>، <code>142</code>، <code>143</code> و <code>144</code> است.
</font>
</p>

<center>
<div dir=rtl style="direction: rtl;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>

| <b>Segment</b> | <b>RFM</b> |
| :---: | :---: |
| Best | <code dir=ltr>144</code> |
| AlmostLost | <code dir=ltr>344</code> |
| LostBigSpenders | <code dir=ltr>444</code> |
| LostCheap | <code dir=ltr>441</code> |
| Loyal | <code dir=ltr>X4X</code> |
| BigSpenders | <code dir=ltr>XX4</code> |

</font>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
    <span style="color:purple"><b>نکته ۱:</b></span>
    در صورتی‌که یک مشتری در هیچ‌یک از گروه‌های جدول قرار نگرفت، گروه او را Normal اعلام کنید.
    <br>
    <span style="color:purple"><b>نکته ۲:</b></span>
    همچنین اگر یک RFM می‌تواند در دو گروه از جدول قرار بگیرد، اولویت با گروهی‌ست که در جدول، <b>بالاتر</b> آمده باشد.
</font>
</p>

```python
df_customer_segments['Segment'] = None # To-Do
df_customer_segments.head()
```

```python
def group_rfm(rfm_value):
    """تابع برای تعیین segment مشتری بر اساس RFM score"""
    rfm_str = str(rfm_value)

    # بررسی شرایط بر اساس اولویت جدول
    if rfm_value == 144:
        return 'Best'
    elif rfm_value == 344:
        return 'AlmostLost'
    elif rfm_value == 444:
        return 'LostBigSpenders'
    elif rfm_value == 441:
        return 'LostCheap'
    elif rfm_str[1] == '4':  # X4X - Loyal customers
        return 'Loyal'
    elif rfm_str[2] == '4':  # XX4 - Big spenders
        return 'BigSpenders'
    else:
        return 'Normal'

# اعمال تابع روی ستون RFM
df_customer_segments['Segment'] = df_customer_segments['RFM'].apply(group_rfm)
df_customer_segments.head()
```
